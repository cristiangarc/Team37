```python
# create a new column, Month, which extracts the month number from the Period column
all_industries["Month"] = all_industries["Period"].str[-2:]
all_industries

# join the Year and Month columns into a new YYYY-MM column
# by casting the columns to strings inside a lambda function, and joining with "-" using the .join method
all_industries["YYYY-MM"] = all_industries.apply(lambda x: "-".join([str(x["Year"]), str(x["Month"])]), axis=1)
all_industries

# create a Date column with the YYYY-MM strings converted to pd.datetime
all_industries["Date"] = pd.to_datetime(all_industries["YYYY-MM"], format="%Y-%m")
all_industries

# convert Date from pd.datetime to Period with frequency set to quarterly
all_industries["Date"] = all_industries["Date"].dt.to_period(freq="Q")
all_industries

# drop the unnecessary columns
all_industries_clean = all_industries.drop(columns=["Series ID", "Period", "Label", "YYYY-MM", "Year", "Month"])
all_industries_clean

# group by Period
all_industries_gob = all_industries_clean.groupby("Date")

# calculate the mean unemployment rate for each quarter
all_industries_quarterly_ue = all_industries_gob.mean()
all_industries_quarterly_ue

all_industries_quarterly_ue.rename(columns = {'Value':'Unemployment Rate'}, inplace = True)
# all_industries_quarterly_ue["Unemployment Rate"] = all_industries_quarterly_ue["Unemployment Rate"].astype(float)
all_industries_quarterly_ue
```

```python
with open('data/Quarterly_Rent_Prices.csv') as f:
    quarterly_rp = pd.read_csv(f)
```

```python
# convert to datetime object
quarterly_rp["Date"] = pd.to_datetime(quarterly_rp["Quarterly"])
quarterly_rp["Date"] = quarterly_rp["Date"].dt.to_period(freq="Q")
quarterly_rp = quarterly_rp.rename(columns={"Date":"Quarter"})
quarterly_rp_copy = quarterly_rp.copy()
quarterly_rp_copy = quarterly_rp_copy.drop('Quarterly', axis=1)
quarterly_rp_copy = quarterly_rp_copy.rename(columns={"United States":"Mean US Rent Price (USD)"})
quarterly_rp_copy = quarterly_rp_copy[["Mean US Rent Price (USD)", "Quarter"]]
quarterly_rp_copy
```

```python
with open("data/Quarterly_House_Price.csv") as f:
    quarterly_hp = pd.read_csv(f)
```

```python
print(quarterly_hp.head(10))
quarterly_hp.dtypes
```

```python
# convert to datetime object
quarterly_hp["Date"] = pd.to_datetime(quarterly_hp["Quarterly"])
quarterly_hp["Date"] = quarterly_hp["Date"].dt.to_period(freq="Q")
quarterly_hp = quarterly_hp.rename(columns={"Date":"Quarter"})
quarterly_hp_copy = quarterly_hp.copy()
quarterly_hp_copy = quarterly_hp_copy.drop('Quarterly', axis=1)
quarterly_hp_copy = quarterly_hp_copy.rename(columns={"United States":"Mean US House Price (USD)"})
quarterly_hp_copy = quarterly_hp_copy[["Mean US House Price (USD)", "Quarter"]]
quarterly_hp_copy
```

```python
# merge unemployment rate dset with house prices dset
uer_hp_merged = pd.merge(left=all_industries_quarterly_ue, right=quarterly_hp_copy, on="Quarter", how="left", suffixes=["_UNEMPLOYMENT_RATE_Q", "_HOUSE_PRICES_Q"])
uer_hp_merged_copy = uer_hp_merged.dropna()
print(uer_hp_merged_copy.head(10))

# pearson correlation between mean quarterly house prices and unemployment rate
print(uer_hp_merged_copy[["Unemployment Rate", "Mean US House Price (USD)"]].corr())

plt.pyplot.scatter(uer_hp_merged_copy["Unemployment Rate"], uer_hp_merged_copy["Mean US House Price (USD)"])
plt.pyplot.title("Mean US House Price (USD) vs Unemployment Rate")
plt.pyplot.xlabel("Unemployment Rate")
plt.pyplot.ylabel("Mean US House Price (USD)")
```

```python
# merge US quarterly rent prices with US quarterly house prices
hp_rent_merged = pd.merge(left=quarterly_hp_copy, right=quarterly_rp_copy, on="Quarter", how="left", suffixes=["_MEAN HOUSE PRICE (USD)", "_MEAN RENT PRICE (USD)"])
hp_rent_merged

# Second merge with UE rate
uer_hp_rent_merged = pd.merge(left=all_industries_quarterly_ue, right=hp_rent_merged, on="Quarter", how="left", suffixes=["_UNEMPLOYMENT RATE", "_MEAN HOUSE/RENT PRICE (USD)"])
uer_hp_rent_merged
```

```python
uer_hp_rent_merged_copy = uer_hp_rent_merged.copy()
uer_hp_rent_merged_copy = uer_hp_rent_merged_copy.dropna()
uer_hp_rent_merged_copy
```

```python
# time series of unemployment, house and rent prices
# plot unemployment rate % over time
uer_2 = uer_hp_rent_merged_copy[["Quarter", "Unemployment Rate"]]
uer_2 = uer_2.set_index("Quarter")
ax = uer_2.plot.line()
ax.set_title("Unemployment rate over time")
ax.set_xlabel("Quarter")
_ = ax.set_ylabel("Unemployment Rate")

# plot US house prices of over time
us_hp = uer_hp_rent_merged_copy[["Quarter", "Mean US House Price (USD)"]]
us_hp = quarterly_hp_copy.set_index("Quarter")
ax2 = us_hp.plot.line()
ax2.set_title("Mean US House Prices Over Time")
ax2.set_xlabel("Quarter")
_ = ax2.set_ylabel("Mean House Price (USD)")

# plot rent prices in US over time
us_rent_prices = uer_hp_rent_merged_copy[["Quarter", "Mean US Rent Price (USD)"]]
us_rent_prices = us_rent_prices.set_index("Quarter")
ax2 = us_rent_prices.plot.line()
ax2.set_title("Mean US Rent Prices Over Time")
ax2.set_xlabel("Quarter")
_ = ax2.set_ylabel("Mean Rent Price (USD)")
```

```python
quarterly_index["Date"] = pd.to_datetime(quarterly_index["Quarterly"])
quarterly_index["Date"] = quarterly_index["Date"].dt.to_period(freq="Q")
quarterly_index_copy = quarterly_index.drop(columns=["Quarterly"])
quarterly_index_copy = quarterly_index_copy.rename(columns={"Date":"Quarter"})
print(quarterly_index_copy.head(10))
```

```python
with open("data/Quarterly_Labor_Movement_Info.csv") as f:
    quarterly_labor = pd.read_csv(f)
```

```python
quarterly_labor.head(10)
```

```python
quarterly_labor.dtypes
```

```python
quarterly_labor["Date"] = pd.to_datetime(quarterly_labor["Quarterly"])
quarterly_labor["Date"] = quarterly_labor["Date"].dt.to_period(freq="Q")
quarterly_labor_copy = quarterly_labor.drop(columns=["Quarterly"])
quarterly_labor_copy = quarterly_labor_copy.rename(columns={"Date":"Quarter"})
print(quarterly_labor_copy.head(10))
```

```python
# merge datasets
# merge unemployment rate, house price, and rent price with quarterly index df
all_merged = pd.merge(left=uer_hp_rent_merged, right=quarterly_index_copy, on="Quarter", how="left", suffixes=["_QUARTER_FIRST", "_QUARTER_INDEX"])

# merge previous set with quarterly_labor df
all_merged = pd.merge(left=all_merged, right=quarterly_labor_copy, on="Quarter", how="left", suffixes=["_QUARTER_SECOND", "_QUARTER_LABOR"])
all_merged.head(10)
```

```python
# drop nas
all_merged_copy = all_merged.dropna()
all_merged_copy = all_merged_copy.reset_index().drop(columns="index")
all_merged_copy.head(10)
all_merged_copy.shape
```

```python
# rename columns
rename_dict = {
    "Unemployment Rate": "unemployment_rate",
    "Mean US House Price (USD)":"mean_US_house_price",
    "Mean US Rent Price (USD)":"mean_US_rent_price",
    "Index.1": "index",
    "Level of activity": "level_of_activity",
    "Momentum": "momentum"
}
all_merged_copy = all_merged_copy.rename(columns=rename_dict)
all_merged_copy.head(10)
```

```python
# linear regression on house prices by rent, labor activity, labor momentum, index
# linear regression on national house prices by index
formula1 = 'mean_US_house_price ~ index'
model1 = smf.ols(formula = formula1, data = all_merged_copy)
fitted1 = model1.fit()
print(fitted1.summary())
```

```python
# linear regression on national house prices by rent price
formula2 = 'mean_US_house_price ~ mean_US_rent_price'
model2 = smf.ols(formula = formula2, data = all_merged_copy)
fitted2 = model2.fit()
print(fitted2.summary())
```

```python
# linear regression on national house prices by rent price, momentum, level of activity
formula2 = 'mean_US_house_price ~ mean_US_rent_price + momentum + level_of_activity'
model2 = smf.ols(formula = formula2, data = all_merged_copy)
fitted2 = model2.fit()
print(fitted2.summary())
```

```python
# linear regression on national house prices by rent price, momentum, level of activity, index
formula2 = 'mean_US_house_price ~ mean_US_rent_price + momentum + level_of_activity + index'
model2 = smf.ols(formula = formula2, data = all_merged_copy)
fitted2 = model2.fit()
print(fitted2.summary())
```

```python
# linear regression on national house prices by rent price, momentum, level of activity, index, unemployment rate
formula2 = 'mean_US_house_price ~ mean_US_rent_price + momentum + level_of_activity + index + unemployment_rate'
model2 = smf.ols(formula = formula2, data = all_merged_copy)
fitted2 = model2.fit()
print(fitted2.summary())
```

```python
# linear regression on national house prices by rent price, momentum, index, unemployment rate
formula2 = 'mean_US_house_price ~ mean_US_rent_price + momentum + index + unemployment_rate'
model2 = smf.ols(formula = formula2, data = all_merged_copy)
fitted2 = model2.fit()
print(fitted2.summary())
```

```python
# linear regression on national house prices by momentum, index, unemployment rate
formula2 = 'mean_US_house_price ~ momentum + index + unemployment_rate'
model2 = smf.ols(formula = formula2, data = all_merged_copy)
fitted2 = model2.fit()
print(fitted2.summary())
```

```python
# linear regression on national house prices by momentum, index, unemployment rate, level of activity
formula2 = 'mean_US_house_price ~ momentum + index + unemployment_rate + level_of_activity'
model2 = smf.ols(formula = formula2, data = all_merged_copy)
fitted2 = model2.fit()
print(fitted2.summary())
```

```python
# linear regression on national house prices by index, unemployment rate, level of activity
formula2 = 'mean_US_house_price ~ index + unemployment_rate + level_of_activity'
model2 = smf.ols(formula = formula2, data = all_merged_copy)
fitted2 = model2.fit()
print(fitted2.summary())
```

```python
# linear regression on national house prices by index, unemployment rate, level of activity, rent prices
formula2 = 'mean_US_house_price ~ index + unemployment_rate + level_of_activity + mean_US_rent_price'
model2 = smf.ols(formula = formula2, data = all_merged_copy)
fitted2 = model2.fit()
print(fitted2.summary())
```
