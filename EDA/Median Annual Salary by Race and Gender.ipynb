```python
import pandas as pd
import csv
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv(r'C:\Users\melinadiaz\Downloads\Median Salary By Race.csv')
```

```python
fig,ax=plt.subplots(1)
plt.xticks(rotation=45) 
```

```python
ax.plot(df[df["Year"]>1999]["Year"],df[df["Year"]>1999]["Total Men"], label="Men", color="blue")
ax.plot(df[df["Year"]>1999]["Year"],df[df["Year"]>1999]["Total Women"], label="Women", color="red")
```

```python
ax.set(xlabel="Year", ylabel="Median Salary",xticks=df[df["Year"]>1999]["Year"])

ax.legend()
fig
```

```python
fig,ax=plt.subplots(1)
plt.xticks(rotation=45) 
columns=['White Women', 'White Men',
       'Black Women', 'Black Men', 'Asian Women', 'Asian Men',
       'Hispanic Women', 'Hispanic Men', 'White, non-Hispanic Women',
       'White, non-Hispanic Men']
for i in columns:
    ax.plot(df[df["Year"]>1999]["Year"],df[df["Year"]>1999][i], label=i)
```

```python
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
ax.set(xlabel="Year", ylabel="Median Salary", xticks=df[df["Year"]>1999]["Year"])
fig
```
